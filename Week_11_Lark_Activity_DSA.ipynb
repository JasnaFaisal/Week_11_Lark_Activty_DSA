```python
#Using the label encoder and one hot encoder 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
labl_enc=LabelEncoder()
a=['school','Mjob', 'Fjob', 'reason', 'guardian','higher','passed', 'sex','address','famsize','Pstatus','schoolsup', 'famsup', 'paid']
for i in np.arange(len(a)):
    data[a[i]]=labl_enc.fit_transform(data[a[i]])
data=pd.get_dummies(data,columns=['activities', 'nursery', 'internet', 'romantic'])    
```

```python
corr_matrix=data.corr()
corr_matrix
```

```python
plt.figure(figsize=(32,20))
sns.heatmap(corr_matrix,vmin=-0.2,vmax=0.7,linewidth=0.3,annot=True)
```

The heatmap and correaltion matrix shows that there is no much high`correlation between the independent variables.

So we combine some of the features,

```python
data['Leisure_time']=data[['freetime', 'goout']].mean(axis=1)
data['Alcohol_cons']= data[[ 'Dalc', 'Walc']].mean(axis=1)
data= data.drop(['freetime','goout','Dalc', 'Walc'],axis=1)
```

```python
data.head()
```

```python
#Creating the model
X= data.drop(['passed'],axis=1)
y= data['passed']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, random_state=25, test_size= 0.2)
model_train_acc=[]
model_test_acc=[]
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf= RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred= rf.predict(X_test)
y_train_pred = rf.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, f1_score
print('Accuracy Score :', accuracy_score(y_test, y_pred))
print('Recall Score :', recall_score(y_test, y_pred))
print('Precision Score :', precision_score(y_test, y_pred))
print('F1 score :', f1_score(y_test, y_pred))
model_train_acc.append(accuracy_score(y_train_pred,y_train))
model_test_acc.append(accuracy_score(y_pred,y_test))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# logistic regression
from sklearn.linear_model import LogisticRegression
lgstc = LogisticRegression()
lgstc.fit(X_train,y_train)
y_pred = lgstc.predict(X_test)
y_train_pred = lgstc.predict(X_train)
```

```python
print('Accuracy Score :', accuracy_score(y_test, y_pred))
print('Recall Score :', recall_score(y_test, y_pred))
print('Precision Score :', precision_score(y_test, y_pred))
print('F1 score :', f1_score(y_test, y_pred))
model_train_acc.append(accuracy_score(y_train_pred,y_train))
model_test_acc.append(accuracy_score(y_pred,y_test))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
#KNN model 
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
knn.fit(X_train,y_train)
y_pred= knn.predict(X_test)
y_train_pred = knn.predict(X_train)
```

```python
print('Accuracy Score :', accuracy_score(y_test, y_pred))
print('Recall Score :', recall_score(y_test, y_pred))
print('Precision Score :', precision_score(y_test, y_pred))
print('F1 score :', f1_score(y_test, y_pred))
model_train_acc.append(accuracy_score(y_train_pred,y_train))
model_test_acc.append(accuracy_score(y_pred,y_test))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# SVM model 
from sklearn.svm import SVC
sv = SVC(kernel = 'rbf', random_state = 0)
sv.fit(X_train, y_train)
y_pred= sv.predict(X_test)
y_train_pred = sv.predict(X_train)
```

```python
print('Accuracy Score :', accuracy_score(y_test, y_pred))
print('Recall Score :', recall_score(y_test, y_pred))
print('Precision Score :', precision_score(y_test, y_pred))
print('F1 score :', f1_score(y_test, y_pred))
model_train_acc.append(accuracy_score(y_train_pred,y_train))
model_test_acc.append(accuracy_score(y_pred,y_test))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
Models = ['svm','Random Forest','Logistic regression','KNN']
model_performance = pd.DataFrame({
    'model':Models,
    'train_acc':model_train_acc,
    'test_acc':model_test_acc
})
model_performance
```
